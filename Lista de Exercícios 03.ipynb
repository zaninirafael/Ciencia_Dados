1. Carregue o dataset para análise, disponível na pasta data, usando as funções de leitura do Pandas.

```python
import numpy as np
import pandas as pd

df = pd.read_csv("data\survey_results_public.csv")

df
```

2. Obtenha informações do Data Frame carregado em memória

```python
print(df.info())
```

3. Selecione todas as linhas com país igual a Brasil

```python
df_brasil = df.loc[df.Country == 'Brazil']

df_brasil
```

4. Os dados do brasil possui dados ausentes (missing data)? Se sim, selecione todas as linhas que contenham algum dado ausente

```python
df_brasil[pd.isna(df_brasil).any(axis=1)]
```

5. Qual o total de valores ausentes para o dataset Brasil?

```python
df_brasil.isna().sum()
```

6. Remova linhas com pelo menos 1 elemento ausente das colunas: 'FormalEducation', 'UndergradMajor', 'Age'.

```python
df_brasil.dropna(subset=['FormalEducation', 'UndergradMajor', 'Age'])

df_brasil
```

7. Crie um subconjunto de dados com as colunas: 

   Respondent, Country, Student, Employment, FormalEducation, UndergradMajor, Exercise, Age

```python
df_analise = df_brasil[['Respondent', 'Country', 'Student', 'Employment', 'FormalEducation', 'UndergradMajor', 'Exercise', 'Age']]

df_analise
```

8. Exiba as contagens de frequência dos valores únicos das colunas: Student, Exercise, Employment, FormalEducation, UndergradMajor e Age

```python
print('Frquência Student:')
print(df_analise['Student'].value_counts())
print('\nFrquência Exercise: ')
print(df_analise['Exercise'].value_counts())
print('\nFrquência Employment: ')
print(df_analise['Employment'].value_counts())
print('\nFrquência FormalEducation: ')
print(df_analise['FormalEducation'].value_counts())
print('\nFrquência UndergradMajor: ')
print(df_analise['UndergradMajor'].value_counts())
print('\nFrquência Age: ')
print(df_analise['Age'].value_counts())
print('\n')

df_analise.describe(include='all')
```
